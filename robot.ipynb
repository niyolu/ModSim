# Modelling
Robots have been used in industrial manufacturing processes for almost 50 years in different areas of application.
An industrial robot is a machine for processing or assembling workpieces that is controlled by computer programs.
Industrial robots are distinguished by the structure of the robot arms, the so-called kinematics.
Robotic arms consist of rigid links that are connected to each other by swivel joints or sliding joints.
The most common robotic arms are 6-axis robots.
They are comparable to the human arm and therefore offer many application possibilities.
Robots have established themselves as an important part of production, particularly in the automotive industry.

## Direct kinematics
The positions of the points $P$, $Q$, $R$, $M$ and $T$ is clearly defined by the two joint angles.
This type of control by the angles $\alpha$ and $\beta$ is called direct kinematics.
Our objective is to derive formulars for the positions of the points $P$, $Q$, $R$, $M$ and $T$ as a function of $\alpha$ and $\beta$.
The parameters $l_1$, $l_2$, $l_3$ and $r$ are also included in these formulas.

The position of point $P$ with respect to the position of point $O$ can be calculated using the length $l_1$ and the angle $\alpha$:

$$
P = O +
l_1 \, 
\left(
\begin{array}{c}
\cos{(\alpha)} \\
\sin{(\alpha)} \\
\end{array}
\right)
$$

In order to calculate the position of point $Q$ we use the angle $\gamma$ as shown in the figure above:

$$
\alpha + \beta - \gamma = \pi
\quad \Longrightarrow \quad
\gamma = \alpha + \beta - \pi
$$

The position of point $Q$ with respect to the position of point $Q$ can be calculated using the length $l_2$ and the angle $\gamma$:

$$
Q
= P
+ l_2 \, 
\left(
\begin{array}{c}
\cos{(\gamma)}\\
\sin{(\gamma)}\\
\end{array}
\right)
$$

By a straightforward calculation we get the positions of the points $R$, $M$ and $T$:  

$$
R = Q
+ l_3 \, 
\left(
\begin{array}{r}
 0\\
-1\\
\end{array}
\right),
\quad
M = R
+ r \, 
\left(
\begin{array}{r}
 0\\
-1\\
\end{array}
\right),
\quad
T = M
+ r \, 
\left(
\begin{array}{r}
 0\\
-1\\
\end{array}
\right)
$$

## Inverse kinematics
Inverse kinematics is exactly the opposite of direct kinematics.
This means that the position of the point $T(T_x \mid T_y)$ is given and the angles $\alpha$ and $\beta$ are calculated.
If we put together all the formulas of direct kinematics, we get:

$$
    \begin{array}{lcl}
        T_x & = & l_1 \, \cos(\alpha) - l_2 \, \cos(\alpha + \beta)\\
		T_y & = & l_1 \, \sin(\alpha) - l_2 \, \sin(\alpha + \beta) - l_3 - 2 \, r
	\end{array}
$$

Now we have to solve these equations according to $\alpha$ and $\beta$.
We could do that through tricky symbolic calculations.
But we choose another alternative.
We solve the equations using a numerical approximation method.
The standard procedure for this purpose is Newton's method for finding roots.
To this end, we must reformulate the equations.
The problem now is:
Find the common roots $\alpha$ and $\beta$ for the functions $f$ and $g$:

$$
	\begin{array}{lclcl}
		f(\alpha,\beta) & = & T_x - l_1 \, \cos(\alpha) + l_2 \, \cos(\alpha + \beta) & = & 0\\
		g(\alpha,\beta) & = & T_y - l_1 \, \sin(\alpha) + l_2 \, \sin(\alpha + \beta) + l_3 + 2 \, r & = & 0
	\end{array}
$$

```python
def parse_function(func_string, coordinates):
    """
    Parses a string into a sympy expression using non-arbitrary coordinate names 
    
    Parameters
    ----------
    `func_string` : list[str]
        The list containing the strings that describe the function

    `coordinates` : list[str]
        The touple containing the ordered coordinates that make up the space of the function

    Returns
    -------
    `expr` : sympy expression
        The resulting generic expression that uses "x_i"-named variables
    """

    expr = list()
    for func_component in func_string:
        for i in range(len(coordinates)):
            func_component = func_component.replace(coordinates[i], "x" + str(i + 1))

        expr.append(sp.sympify(func_component))

    return expr



def jacobi(f):
    """
    Calculates the jacobi matrix of a function `f` : R^n -> R^m and
    
    Parameters
    ----------
    `f` : list[sympy expression]
        The list containing said function, requires generic non-constant names

    Returns
    --------
    `J_sp` : Sympy Matrix
        Symbolic Jacobi Matrix of `f`

    `J_np` : Lambdified Matrix
        Matrix containing Jacobi Lambdas
    """


    # first we need to sperate variables and constants for differentiation
    variables = set()
    constants = set()

    # seperate atoms using string representation and regex
    p = re.compile("^x[0-9]*$")

    for f_component in f:
        atoms = f_component.atoms(sp.Symbol)
        for atom in atoms:
            if p.match( str(atom) ):
                variables.add(atom)
            else:
                constants.add( str(atom) )

    m = len(f)
    n = len(variables)

    # sort constants alphabetically in order to ensure non-arbitrary value passing
    constants = sp.sympify( sorted(list(constants)) )

    # generate uniform and ordered coordinates
    coordinates = [sp.sympify( "x" + str(i) ) for i in range(1, n + 1)]

    jacobi_list = list()

    # calculate gradients and insert them into matrix rows
    for i in range(m):
        f_component = f[i]
        gradient = list()

        for j in range(n):

            # differentiate the component function with respect to the coordinates
            dif = sp.diff( f_component, coordinates[j] )
            gradient.append(dif)

        jacobi_list.append(gradient)

    # create sympy and lambdified numpy versions of the jacobi matrix
    J_sp, J_np = sp.Matrix( jacobi_list ), sp.lambdify( coordinates + constants, jacobi_list, modules = "numpy" )
    return J_sp, J_np



# driver code
if __name__ == "__main__":
    constants = {"Tx" : 0, "Ty" : 0, "l1" : 6.0, "l2" : 6.0, "l3" : 1.0, "r" : 0.5}
    function_string = [ "Tx - l1 * cos(alpha) + l2 * cos(alpha + beta)",
                        "Ty - l1 * sin(alpha) + l2 * sin(alpha + beta) + l3 + 2 * r"]
    coordinates = ["alpha", "beta"]
    values = (pi/3, pi/2)

    expr = parse_function(function_string, coordinates)
    J_sp, J_np = jacobi(expr)
    J = sp.Matrix(J_np(*values, *constants.values()))

    print(f"Calculated the Jacobi Matrix of:")
    display(sp.Matrix(expr))
    print("as: ")
    display(J_sp)
    print(f"Evaluated at {values}:")
    display(J)
```

## Collision detection
One of the most important tasks when simulating robots is to make sure that there are no collisions.
Collisions can occur in different forms.
The object could collide with the obstacle, with an arm, or with the ground.

Functions that intersect straight lines with straight lines or with circles form the mathematical basis of collision checks.

We take a closer look at the calculation of the intersection of a straight line with a circle.
We represent a straight line with the constants $a$, $b$ and $c$ with the equation

$$
a \, x + b \, y = c
$$

and a circle with midpoint $M(x_0 \mid y_0)$ and radius $r$ with the equation

$$
(x - x_0)^2 + (y - y_0)^2 = r^2
$$

We use the transformation

$$
\tilde{x} = x - x_0, \quad \tilde{y} = y - y_0 
$$

to simplify the equation of the circle to

$$
\tilde{x}^2 + \tilde{y}^2 = r^2
$$

With this transformation the equation for the straight line is given by

$$
	a (\tilde{x} + x_0) + b (\tilde{y} + y_0) = c
    \quad \Longrightarrow \quad
    a \, \tilde{x} + b \, \tilde{y} = 
    \underbrace{c - a \, x_0 - b \, y_0}_{\displaystyle d} 
$$

In order to simplify the notation we use the shortcut

$$
d = c - a \, x_0 - b \, y_0
$$

In case of $b \neq 0$ we can solve this equation for $\tilde{y}$

$$
\tilde{y} = \frac{d - a \tilde{x}}{b}
$$

Inserting this in the equation for the circle results in the quadratic equation

$$
\tilde{x}^2 + \left( \frac{d - a \tilde{x}}{b}\right)^2 = r^2
$$

This can be written as

$$
\underbrace{\left(1 + \frac{a^2}{b^2} \right)}_{\displaystyle A} \, \tilde{x}^2 \, + \,
\underbrace{\frac{-2 \, a \, d}{b^2}}_{\displaystyle B} \, \tilde{x} \, + \, 
\underbrace{\frac{d^2}{b^2} - r^2}_{\displaystyle C} = 0
$$

After multiplication with $b^2$ this results in

$$
\underbrace{\left(a^2 + b^2 \right)}_{\displaystyle A} \, \tilde{x}^2 \, - \,
\underbrace{2 \, a \, d}_{\displaystyle -B} \, \tilde{x} \, + \, 
\underbrace{d^2 - r^2 \, b^2}_{\displaystyle C} = 0
$$

With the shortcuts

$$
A = a^2 + b^2, \quad
B = -2 \, a \, d, \quad 
C = d^2 - r^2 \, b^2
$$

the problem ends up in a quadratic equation for $\tilde{x}$

$$
A \, \tilde{x}^2 + B \, \tilde{x} + C = 0
$$

As expected, the intersection problem of a stright line with a circle can either have two solutions, one solution, or no solution at all.

In case of $b = 0$ we can proceed similarly.
Since in this case we can guarantee $a \neq 0$.

```python
def robotDirectKinematics(alpha, beta):
    """
    Calculates the resulting positions of the arm angles

    Parameters
    ----------
    `alpha`, `beta` : float
        The angles describing the robotic arm

    Returns
    --------
    `O`, `P`, `Q`, `R` : numpy array
        The positions of the robot arm sections/joints,
        Origin to End

    `M`, `T` : numpy array
        The positions of the object midpoint and bottom,
        respectively
    """
    gamma = alpha + beta - pi
    O = np.array([0, 0])
    P = O + l1 * np.array([cos(alpha), sin(alpha)])
    Q = P + l2 * np.array([cos(gamma), sin(gamma)])

    R = Q + l3 * np.array([0, -1])
    M = R + r * np.array([0, -1])
    T = M + r * np.array([0, -1])

    return P, Q, R, M, T
```

<b> Generic and specific implementations of multi-dimensional Newton Raphson Algorithm:</b>

```python
class convergenceError(Exception):
    """
    Exception class for raising Errors on non-convergent Algorithms

    Attributes
    ----------
    `n` : int
        Describes the iteration limit at which the Algorithm stopped unsuccesfully
    """
    def __init__(self, n):
        self.n = n

    def __str__(self):
        return f"Numeric Algorithm did not converge after {self.n} steps"
```

```python
def multiNewton_generic(f, x_k, n = 10, debug=False):
    """
    Newton-Raphson Method for approximating a root of a function

    Parameters
    ----------
    `f` : sympy Matrix
        Matrix of multivariable functions

    `x_k` : numpy array
        Initial guess of the solution

    `n` : int, optional (default = 10)
        Iteration limit 

    `debug` : bool, optional (default = False)
        print debug information while running the method

    Returns
    -------
    `x_k` : numpy array
        approximated root

    Raises
    ------
    convergenceError
        If the iteration limit was reached with no result
    """

    Y = sp.Matrix([x, y])
    J = f.jacobian(Y)

    J = sp.lambdify([x, y], J, modules="numpy")
    f = sp.lambdify([x, y], f)

    for i in range(n):

        dx = np.linalg.solve( J(x_k[0], x_k[1]), -f(x_k[0], x_k[1]) )
        dx = np.array([dx[0][0], dx[1][0]])
        x_k = x_k + dx

        deviation = max(abs(dx))

        if debug:
            print(f"x_{i} = ", x_k)
            print(f"Deviation = {deviation}")

        if deviation < np.finfo(float).eps:
            print("Solution:")
            display(x_k)
            return x_k

    raise convergenceError(n)
```

```python
def multiNewton(T, alpha, beta, n=10, debug=False):
    """
    Newton-Raphson Method for approximating a solution to

    Parameters
    ----------
    `T` : numpy array
        Position of the object-bottom we try to solve for

    `alpha`, `beta` : float
        Initial guesses for the angles

    `n` : int, optional (default = 10)
        Iteration limit 

    `debug` : bool, optional (default = False)
        print debug information while running the method

    Returns
    -------
    `alpha`, `beta` : numpy array
        approximated root

    Raises
    ------
    convergenceError
        If the iteration limit was reached with no result
    """

    Tx, Ty, Alpha, Beta, L1, L2, L3, R = sp.symbols("Tx Ty alpha beta l1 l2 l3 r")
    parameters = [Alpha, Beta, Tx, Ty, L1, L2, L3, R]
    values = [T[0], T[1], l1, l2, l3, r]

    f = sp.Matrix([
                    [Tx - L1 * sp.cos(Alpha) + L2 * sp.cos(Alpha + Beta)],
                    [Ty - L1 * sp.sin(Alpha) + L2 * sp.sin(Alpha + Beta) + L3 + 2 * R]
    ])

    J = f.jacobian([Alpha, Beta])

    if debug: display(J)

    J = sp.lambdify(parameters, J, modules="numpy")
    f = sp.lambdify(parameters, f, modules="numpy")
    x = np.array([alpha, beta])

    for i in range(n):
        dx = np.linalg.solve( J(*x, *values), -f(*x, *values).flatten())
        x = x + dx
        deviation = max(abs(dx))

        # the algorithm did not always converge using np.finfo(float).eps on 3.7.6 64Bit Conda
        # so we used a pre-defined constant. It worked on collabs though.
        if deviation < 1e-15:
            x = tuple(x)
            if debug: print("Solution after %d steps: alpha = %f, beta = %f" % (i+1, *x))
            return x

    raise convergenceError(n)
```

```python
def robotInverseKinematics(T, alpha, beta):
    """
    wraps multivariable Newton method to calculate a path towards a given position
    """
    try:
        return multiNewton(T, alpha, beta, n=50, debug=False)
    except convergenceError as err:
         print("robotInverseKinematics failed: " + str(err))
```

## Path planning

The motion will start at position $(x_0 \mid 0)$ and end at position $(x_1 \mid 0)$.

```python
def rectRobotPath(x0, x1, y, n):
    """
    Path generation along a rectangle

    Parameters
    ----------
    `x0`, `x1` : float
        x-value of the starting and end points, respectively
    
    `y` : float
        height of the box we are pathing around

    `n` : int
        number of path-points

    Returns
    -------
    `path` : numpy matrix[xi, yi]
        contains the resulting path-points
    
    """
    
    x0 = x0
    x1 = x1
    y1 = y
    y0 = 0
    
    dy = y1 - y0
    dx = x1 - x0
    
    n -= 1 if n % 3 else 0
    
    nx = np.ceil(n/3)
    ny = n - nx
    
    sx = dx / nx
    sy = 2 * dy / ny
        
    xstep = np.array([sx, 0])
    ystep = np.array([0, sy])
    
    last = np.array([x0, y0])
    
    path = np.array([last])
    
    err = 10e-5
    
    for i in range(n):
        current = np.copy(last)
        
        if not(current[0] <= x1 + err and current[0] > x1 - err):
            if not (current[1] <= y1 + err and current[1] > y1 - err):
                current += ystep
            else:
                current += xstep
        else:
            current -= ystep
            
        add = np.array([current])
        path = np.concatenate((path, add))
        last = current

    return path
```

```python
def polynomialRobotPath(x0, x1, y, n, debug=False):
    """
    Path generation along a 3rd-order polynomial P(x)
   
    Parameters
    ----------
    `x0`, `x1` : float
        x-value of the starting and end points, respectively
    
    `y` : float
        height of the box we are pathing around
    `n` : int
        number of path-points

    Returns
    -------
    `path` : numpy matrix[xi, yi]
        contains the resulting path-points
    """
    
    global w
    global d
    

    # we path around 4 points:

    px = np.array([
        x0,         #starting-point
        d - w/2,    # left obstacle edge
        d + w/2,    # right obstacle edge
        x1          # end-point
    ])

    py = np.array([
        0,          # starting-point
        y,          # left obstacle edge
        y,          # right obstacle edge
        0           # end-point
    ])

    # substituting those 4 points into our P(x) leaves us with 4 equations,
    # that we can rewrite into a matrix multiplicaiton
    # V(px) * c = py, where V(px) is the Vandermonde matrix 
    V = np.vander(px, increasing=True)

    # we solve the matrix equation for the constants c and build our polynomial
    c = np.linalg.solve(V, py)

    # somehow if we use standard x the code crashes on repeated calls to this function
    X = sp.symbols("X")

    # we now have our polynomial
    p = c[0] + c[1] * X + c[2] * X**2 + c[3] * X**3
    
    if debug: display(p)

    p = sp.lambdify(X, p)

    stepx = (x1 - x0)/(n-1)
    pathx = [stepx * i + x0 for i in range(n)]
    pathy = [p(x) for x in pathx]
    path = np.array([[pathx[t], pathy[t]] for t in range(len(pathx))])
  
    return path
```

```python
def robotPath(x0, x1, y, n):
    """
    Path generation from point (`x0`, 0) to point (`x1`, 0)
   
    Parameters
    ----------
    `x0`, `x1` : float
        x-value of the starting and end points, respectively
    
    `y` : float
        height of the box we are pathing around
    `n` : int
        number of path-points

    Returns
    -------
    `path` : numpy array[xi, yi]
        contains the resulting path-points
    
    """

    global d

    # symmetric case
    if d-x0 == x1-d:
        return polynomialRobotPath(x0, x1, y, n)

    # asymmetric case
    else:
        return rectRobotPath(x0, x1, y, n)
```

```python
def solve_quadratic(A, B, C):
    """
    Solves a quadratic equation : `A`x² + `B`x + `C` = 0
   
    Parameters
    ----------
    `A`, `B`, `C` : float
        constants of the equation

    Returns
    -------
    `x` : list[float]
        the solution that is either empy, two or on
    """
    
    discr = B**2 - 4*A*C

    # two solutions
    if discr > 0:
        x1 = (-B - np.sqrt(discr)) / (2*A)
        x2 = (-B + np.sqrt(discr)) / (2*A)
        x = [x1, x2]
    
    # one solution
    elif discr == 0:
        x = [ -B / (2*A) ]

    # complex solution = no solution
    else:
        x = []

    return x


def intersectLineCircle(a,b,c,M,r):
    """
    Caclulctes the circle-line-intersection for:
    
    line:   `a`*x + `b`*y = `c`
    circle: (x - `M[0]`)² + (y - `M[1]`)² = `r`²
   
    Parameters
    ----------
    `a`, `b`, `c` : float
        coefficients of the straight line
    
    `M` : list[float]
        Midpoint of the circle
    
    `r` : float
        radius of the circle

    Returns
    -------
    `S` : numpy array([xi, yi])
        contains resulting intersection,
        either one two or empty
    """
    # no solution array
    no_S = np.array([[]])

    # horizonal edge case
    if abs(a) < np.finfo(float).eps:
        y = c/b
        z = np.sqrt(-(y-M[1])**2 + r**2)

        # no intersection
        if np.isnan(z):
            S = no_S

        # one intersection:
        elif abs(z) < np.finfo(float).eps:
            S = np.array([[M[0], y]])

        # two intersections
        else:
            S = np.array([[M[0]-z, y], [M[0]+z, y]])

    # vertical edge case
    elif abs(b) < np.finfo(float).eps:
        x = c/a
        z = np.sqrt(-(x-M[0])**2 + r**2)

        #no intersection
        if np.isnan(z):
            S = no_S

        # one intersection
        elif abs(z) < np.finfo(float).eps:
            S = np.array([[x, M[1]]])

        # two intersections
        else:
            S = np.array([[x, M[1]-z], [x, M[1]+z]])


    # general case
    else:
        d = c - a*M[0] - b*M[1]
        A = a**2 + b**2
        B = -2 * a * d
        C = d**2 - (b**2 * r**2)

        # resubstitute
        AA = A
        BB = -2*A*M[0]+B
        CC = A*M[0]**2 - B*M[0] + C

        sx = solve_quadratic(AA, BB, CC)

        # no intersection
        if len(sx) == 0:
            S = no_S

        # one or two intersections
        else:
            S = np.array([[x, (c - a*x)/b] for x in sx])

    return S
```

# Evaluation
Based on the animation, we get the impression that our modeling and simulation deliver reliable results.
However, when we take a closer look, we still see problems.
When evaluating our simulations, two different aspects play a role: verification and validation.
Verification and validation are not the same thing, although they are often confused.
## Verification
Verification checks that the specifications are correctly implemented.
It gives an answer to the question:

__Are we implementing the simulation right?__

To verify, we have to take into account, that Newton's method does not always calculate the correct angles.
## Validation
Validation checks that the simulation fullfills the needs and goals of the stakholders.
It gives an answer to the question:

__Are we implementing the right simulation?__

If we compare our model with reality, we would discover differences.
We concentrated only on the pure motion sequence and disregarded all physical quantities such as mass, forces and friction.

```python
def collision(path):
    """
    Checks a `path` for collisions
    
    Parameters
    ----------
    `path` : numpy array[[xi, yi]]
        the path we want to check

    Returns
    -------
    True/False : boolean
        wether or not a collision has been detected
    """
    
    lines = [
        {"a" : 1.0, "b" : 0.0, "c" : d - w/2},    # obstacle left edge
        {"a" : 1.0, "b" : 0.0, "c" : d + w/2},    # obstacle right edge
        {"a" : 0.0, "b" : 1.0, "c" : h}           # obstacle top edge
    ]

    for P in path:
        M = P + np.array([0, r])
        for line in lines:
            S = intersectLineCircle(line["a"], line["b"], line["c"], M, r)

            if len(S[0]) < 1: continue

            # horizontal case
            if line["a"] == 0:
                for SS in S:
                    if SS[0] >= d - w/2 and SS[0] <= d + w/2:
                        return True

            # vertical case
            elif line["b"] == 0:
                for SS in S:
                    if SS[1] <= h:
                        return True
                
    # no collision
    return False


def optimize_path(x0, x1, n, start = 1.0, step = 1.0):
    """
    Optimizes a path under the condition of not crashing
   
    Parameters
    ----------
    `x0`, `x1` : float
        start and end x-positions, respectively
    
    `n` : int
        number of path points
    
    `start` : float
        starting-height

    `step` : float
        height increment

    Returns
    -------
    `current_path` : numpy array[[xi, yi]]
        resulting path
    """


    height = start

    # initial path
    current_path = robotPath(x0, x1, height, n)
    
    # recompute path until there are no collisions
    while(collision(current_path)):
        height += step
        current_path = robotPath(x0, x1, height, n)
    
    return current_path
```
